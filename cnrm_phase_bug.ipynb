```python
iris.FUTURE.netcdf_promote=True
ppp=iris.load('/project/champ/data/cmip5/output1/CNRM-CERFACS/CNRM-CM5/rcp85/6hr/atmos/6hrPlev/r1i1p1/v20130101/psl/psl_6hrPlev_CNRM-CM5_rcp85_r1i1p1_202701010600-202801010000.nc')
uuu=iris.load('/project/champ/data/cmip5/output1/CNRM-CERFACS/CNRM-CM5/rcp85/3hr/atmos/3hr/r1i1p1/v20130101/uas/uas_3hr_CNRM-CM5_rcp85_r1i1p1_202601010300-203101010000.nc')
vvv=iris.load('/project/champ/data/cmip5/output1/CNRM-CERFACS/CNRM-CM5/rcp85/3hr/atmos/3hr/r1i1p1/v20130101/vas/vas_3hr_CNRM-CM5_rcp85_r1i1p1_202601010300-203101010000.nc')
pp=ppp[0]
uu=uuu[0]
vv=vvv[0]
p=pp[147]
u=uu[3215]
v=vv[3215]
```

```python
print (p.coord('time')).points
print (u.coord('time')).points
print (v.coord('time')).points
```

So p,u, and v are all nominally the same time. Just check that they are not using a different basis time:

```python
from iris.util import unify_time_units
unify_time_units( [ p, u, v ])

print (p.coord('time')).points
print (u.coord('time')).points
print (v.coord('time')).points
```

The bounds are slightly different because pressure is 6-hourly, whereas u and v are 3-hourly:

```python
print (p.coord('time')).bounds
print (u.coord('time')).bounds
print (v.coord('time')).bounds
```

```python
qplt.pcolormesh(v)
qplt.contour(p)
qplt.contour(p, np.arange(95000,100000, 400 ), colors='black')
plt.gca().coastlines()
plt.show()
```

It's not easy to see the problem at global scale, but if you do this in an ipython session or if you can use Phil's https://pelson.github.io/2014/nbagg_backend/

For example, here is a zoom showing wind vectors:

```python
p2 = p.intersection(longitude=[-35., 15.], latitude=[45., 75.])
u2 = u.intersection(longitude=[-35., 15.], latitude=[45., 75.])
v2 = v.intersection(longitude=[-35., 15.], latitude=[45., 75.])
x2=u2.coord('longitude').points
y2=u2.coord('latitude').points
step=1
qplt.contour(p2)
qplt.contour(p2, np.arange(95000,100000, 400 ), colors='blue')
plt.gca().coastlines()
plt.quiver(x2[::step], y2[::step], u2.data[::step, ::step], v2.data[::step, ::step], scale=300.)
plt.show()
```
